```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ablation.csv")

x_labels = data["model"].astype(str)
x = np.arange(len(x_labels))

def plot_metric(metric_name, metric_values, color):
    plt.plot(x, metric_values, marker='o', label=metric_name, color=color)

plt.figure(figsize=(10, 6))
plot_metric("LLM Accuracy", data["llm_acc"], "blue")
plot_metric("VQA MQM", data["vqa_mqm"], "green")
plot_metric("TitleVQA", data["titlevqa"], "red")
plot_metric("RefVQA", data["refvqa"], "purple")

plt.xticks(x, x_labels)
plt.xlabel("Model Size")
plt.ylabel("Accuracy (%)")
plt.title("Performance of InternVL-2.5 with increasing model size")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
