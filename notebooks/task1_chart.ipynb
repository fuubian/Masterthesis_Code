```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("task1_results.csv")
    
df.set_index('model', inplace=True)
df.T.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Model Performance Across Metrics')
plt.legend(title='Models')
plt.grid(True)
plt.show()
```
