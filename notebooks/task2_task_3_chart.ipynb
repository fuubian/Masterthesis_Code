```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('task23.csv', sep=';', header=None, names=['Model', 'TitleVQA', 'RefVQA'])

df['Average'] = (df['TitleVQA'] + df['RefVQA']) / 2
df = df.sort_values(by='Average', ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(df['Model']))
width = 0.4

ax.bar(x - width/2, df['TitleVQA'], width, label='TitleVQA', color='#d62728')
ax.bar(x + width/2, df['RefVQA'], width, label='RefVQA', color='#1f77b4')

ax.set_xlabel("Model")
ax.set_ylabel("Accuracies")
ax.set_title("Comparison of Models")
ax.set_xticks(x)
ax.set_xticklabels(df['Model'], rotation=45, ha="right")
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
